## Clean Up the Data!
The first step in this is figuring out what we need to clean. I have found in practice, that you find things you need to clean as you perform operations and get errors. Based on the error, you decide how you should go about cleaning the data.

#### Drop the Unrelated/Blank column

```python
df.drop(columns=['Status','unnamed1'],axis=1,inplace= True)# Using the drop function we delete  two columns ans save the changes in our data set
```

# EDA : Exploratory Data Analysis

### Question : Count the Data Gender Wise ?

```python
ax = sns.countplot(x='Gender',data=data,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

### Question : Count the Gender data with Total Amount ?

```python
sales_gender = data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount')
sales_gender
```

```python
sales_gender = data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gender,hue='Gender')
  
plt.show()
```


### Insight : *From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Question : Count the data according to Age_Group ?

```python
ax= sns.countplot(x='Age Group',data=data,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

### Question : Count the Total Sale By Age Group ?

```python
sales_gender = data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_gender,palette='ocean')
  
plt.show()
```

### Insight : *From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*


### Question :  Find the Total number of Orders from top 10 States ?

```python
sales_state = data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})

sns.barplot(x='State',y='Orders',data=sales_state,palette='Dark2')
  
plt.show()
```

### Question : find the  total amount/sales from top 10 states ?

```python
sales_state = data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})

sns.barplot(x='State',y='Amount',data=sales_state,palette='Dark2')
  
plt.show()
```

### Insight : *From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Question : Count the customer according to Marital_Status ?

```python
ax= sns.countplot(x='Marital_Status',data=data)


sns.set(rc={'figure.figsize':(7,4)})
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

###  Question : Find the Total sales according to marital_status by genderwise ?

```python
sales_marital=data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data = sales_marital, x = 'Marital_Status',y= 'Amount', hue='Gender')

plt.show()
```

### Insight : *From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*


### Question : Find the count of Occupation data by Gender Wise ?

```python
sns.set(rc={'figure.figsize':(20,5)})

ax= sns.countplot(data=data,x="Occupation",hue='Gender',palette='Dark2')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

### Question : Find the Total Sales by Occupation wise ?

```python
sales_occupation=data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data = sales_occupation, x = 'Occupation',y= 'Amount',palette='ocean')

sns.set(rc={'figure.figsize':(20,5)})

plt.show()
```

### Insight 4 : *From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*


### Question : Find the Count of product Category ?

```python
ax = sns.countplot(x='Product_Category',data=data)

for bars in ax.containers:
    ax.bar_label(bars)


sns.set(rc={'figure.figsize':(20,5)})

plt.show()
```

### Question : Find the product Category by amount wise ?

```python
sales_Product_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_Product_category, x = 'Product_Category',y= 'Amount')

plt.show()
```

### Insight 5 : *From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*
